## Cleaning the data

```python
import pandas as pd
```

```python
data = pd.read_csv('data.csv', header=None)
```

```python
data
```

```python
data.shape
```

```python
data_col = data.iloc[0].str.split(",").tolist()[0]
```

```python
data_col
```

```python
data
```

```python
rows = data.iloc[1]
rows
```

```python
data = data.iloc[1:]
```

```python
data
```

```python
rows_list = data.values.tolist()
```

```python
rows_list
```

```python
cleaned = [item for sublist in rows_list for item in sublist]
sep = [element.split(",") for element in cleaned]
```

```python
sep
```

```python
cleaned_df = pd.DataFrame(sep, columns=data_col)
```
